```python
# Zadanie 1
def prime(n):
    if n == 1:
        return False

    primes = []
    for i in range(1, n+1):
        is_primary = True
        for p in primes:
            if i % p == 0 and p != 1:
                is_primary = False
        if is_primary:
            primes.append(i)
            
    if n == primes[len(primes)-1]:
        return True
    else:
        return False


def select_primes(x):
    primes = []
    for number in x:
        if prime(number):
            primes.append(number)

    return primes


print(prime(3))
print(prime(4))
print(prime(49))

print(select_primes([3, 6, 11, 25, 19]))
```

```python
# Zadanie 2 e)
import statistics
for miasto in ['Gdansk', 'Poznan', 'Szczecin']:
    mean = statistics.mean(miasta[miasto])
    sd = statistics.stdev(miasta[miasto])

    for i in range (len(miasta[miasto])):
        miasta[miasto][i] = (miasta[miasto][i]-mean)/sd

    print('\nUstandaryzowany', miasto)
    mean = round(statistics.mean(miasta[miasto]))
    print("rednia", mean)
    sd = round(statistics.stdev(miasta[miasto]))
    print("Odchylenie standardowe", sd)
```

```python
# Zadanie 2 f)
miasta = pd.read_csv('miasta.csv')

for miasto in ['Gdansk', 'Poznan', 'Szczecin']:
    minimum = min(miasta[miasto])
    maximum = max(miasta[miasto])

    for i in range (len(miasta[miasto])):
        miasta[miasto][i] = (miasta[miasto][i]-minimum)/(maximum-minimum)

    print('\nZnormalizowany', miasto)
    mean = round(statistics.mean(miasta[miasto])*100)/100
    print("rednia", mean)
    sd = round(statistics.stdev(miasta[miasto])*100)/100
    print("Odchylenie standardowe", sd)
```
